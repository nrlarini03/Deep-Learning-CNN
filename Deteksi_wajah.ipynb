```python

import face_recognition
import cv2
import matplotlib.pyplot as plt

# Data: nama dan usia
data_wajah = [
    {"file": "NurulArini.png", "nama": "Nurul", "usia": 22},

]

# Loop untuk menampilkan hasil deteksi tiap wajah
for data in data_wajah:
    img = face_recognition.load_image_file(data["file"])
    face_locations = face_recognition.face_locations(img)

    # Konversi ke format BGR untuk OpenCV
    img_cv = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

    for (top, right, bottom, left) in face_locations:
        cv2.rectangle(img_cv, (left, top), (right, bottom), (0, 255, 0), 2)
        cv2.putText(img_cv, f"{data['nama']}, {data['usia']} th", (left, top - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

    # Tampilkan hasil
    plt.imshow(cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title(f"Wajah: {data['nama']} (Usia: {data['usia']} th)")
    plt.show()
```

```python
# Install library (jalankan sekali saja)
!pip install opencv-python matplotlib

# Import library
import cv2
import matplotlib.pyplot as plt
import os

# === 1️⃣ Hubungkan Google Drive ===
from google.colab import drive
drive.mount('/content/drive')

# === 2️⃣ Tentukan path folder tempat gambar disimpan ===
# Ubah path ini sesuai lokasi folder di Drive kamu
folder_path = "/content/drive/MyDrive/Wajah/"

# === 3️⃣ Data nama dan usia ===
data_wajah = [
    {"file": "NurulArini.png", "nama": "Nurul", "usia": 22},
    {"file": "adis1.jpg", "nama": "Adis", "usia": 17},
    {"file": "adis2.jpg", "nama": "Adis", "usia": 25},
    {"file": "azel.jpg", "nama": "Azel", "usia": 40},
    {"file": "lulu.jpeg", "nama": "Lulu", "usia": 31},
    {"file": "lulu1.jpg", "nama": "Lulu", "usia": 56},
    {"file": "tania.jpg", "nama": "Tania", "usia": 12},
    {"file": "tania4.jpg", "nama": "Tania", "usia": 40},
    {"file": "tania2.jpg", "nama": "Tania", "usia": 26},
    {"file": "tania3.jpg", "nama": "Tania", "usia": 18},
]

# === 4️⃣ Load classifier deteksi wajah bawaan OpenCV ===
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# === 5️⃣ Loop untuk membaca gambar dari Drive dan menampilkan hasil deteksi ===
for data in data_wajah:
    file_path = os.path.join(folder_path, data["file"])

    if not os.path.exists(file_path):
        print(f"❌ File tidak ditemukan: {file_path}")
        continue

    # Baca gambar
    img = cv2.imread(file_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Deteksi wajah
    faces = face_cascade.detectMultiScale(gray, 1.1, 5)

    # Gambar kotak dan teks nama + usia
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(img, f"{data['nama']}, {data['usia']} th", (x, y-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (36, 255, 12), 2)

    # Tampilkan hasil
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title(f"Wajah: {data['nama']} (Usia: {data['usia']} th)")
    plt.show()
```
